```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df['stator_winding']
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['stator_winding'], bins=30)
plt.xlabel("Stator Winding Temperature")
plt.ylabel("Frequency")
plt.title("Distribution of Motor Temperature")
plt.show()
```

```python
corr = df.corr()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(corr, cmap='coolwarm')
plt.title("Correlation Heatmap of Motor Parameters")
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.scatter(df['motor_speed'], df['stator_winding'], alpha=0.3)
plt.xlabel("Motor Speed")
plt.ylabel("Stator Winding Temperature")
plt.title("Motor Speed vs Motor Temperature")
plt.show()
```
